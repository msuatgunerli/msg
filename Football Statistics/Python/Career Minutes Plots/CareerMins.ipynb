```python
df = pd.read_excel('GS Playing Times.xlsx')
df = df.fillna(0)
df = pd.concat([df.iloc[::, :2], df.iloc[::, 2:].cumsum(axis=1)], axis=1) # COMMENT OUT TO PLOT YEARLY DATA
df = df.T

df.columns = df.iloc[0] # rename columns using the 1st row
df = df[1:] # remove the first row
df_birthdates = df.iloc[:1, ::] # get the birthdates for each column
#birth_dates = df_birthdates.values.tolist()
df = df[1:] # remove the first row
df['Dates'] = df.index # create a dates column using the timestamps in the index
#df.reset_index(drop=True, inplace=True) # drop the index column

player_names = list(df.columns[:-1])
df.head()
```

```python
def plotParams():
    plt.title('Age vs. Cumulative Minutes Played (Club)', color = 'white')
    plt.xlabel('Age', color='white')
    plt.ylabel('Cumulative Minutes Played (Club)', color='white')
    ax.set_facecolor('#121212')
    fig.set_facecolor('#121212')
    ax.tick_params(color='white', labelcolor='white')
    for spine in ax.spines.values():
        spine.set_edgecolor('white')
```

```python
fig, ax = plt.subplots(figsize = (14,7))
x = list(df.reset_index()['Dates'])
x[:] = [number.year for number in x]
y = df[player_names]
ax.plot(x, y, marker='o')
ax.legend(y, loc='upper left')

plotParams()
plt.title('Year vs. Cumulative Minutes Played (Club)', color='white')
plt.xlabel('Year', color='white')
plt.xticks(x)
fig.tight_layout()
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

for name in player_names:
    mins = df[name].loc[(df[[name]] != 0).any(1)]
    dob = df_birthdates[name].loc[(df_birthdates[[name]] != 0).any(1)].values
    ages = list(mins.index.values)
    ages[:] = [number - dob for number in list(mins.index)]
    agelist = []
    for age in enumerate(ages):
        agelist.append(((ages[age[0]][0].days/360)))
    ax.plot(agelist, mins, marker = 'o')
    ax.legend(player_names, loc='upper left')
    
plotParams()
plt.xticks(range(16, 41))
fig.tight_layout()
fig.show()
```
