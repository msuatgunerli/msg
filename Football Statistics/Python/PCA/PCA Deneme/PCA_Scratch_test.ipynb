```python
%matplotlib widget
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn import datasets
import numpy as np
from PCA_Scratch import PCA
```

```python
data = datasets.load_iris()
X = data.data 
y = data.target
```

```python
pca = PCA(3)
pca.fit(X)
X_proj = pca.transform(X)

print(X.shape)
print(X_proj.shape)
```

```python
x1 = X_proj[:,0]
x2 = X_proj[:,1]
x3 = X_proj[:,2]
```

```python
%matplotlib inline

fig = plt.figure()
plt.scatter(x1, x2, c=y, edgecolor = 'none', alpha = 0.8, cmap = plt.get_cmap('jet',3))
plt.title('2 Principal Components')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
plt.show()
```

```python
%matplotlib widget
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, x3, c=y, edgecolor='none', alpha=0.8, cmap=plt.get_cmap('jet', 3))

plt.title('3 Principal Components')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
```
