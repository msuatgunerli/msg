```python
from functools import reduce
#import highlight_text
import pandas as pd
%matplotlib widget
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
#import scipy as sp
#from pylab import text
#from soccerplots.radar_chart import Radar
#!pip install highlight_text == 0.0.5
```

```python
Games_Played = 11
Mins_Played = Games_Played * 90
Team_Name = 'Galatasaray'
```

```python
df = pd.read_excel('Galatasaray.xlsx', sheet_name='21-22')
df['Mins'] = df['Unnamed: 8']
df['Nationality'] = df['Nat.']
df.dropna(axis='rows', inplace=True, subset=['Mins'])
df.reset_index(drop=True, inplace = True)
df = df[['player','Age', 'Nationality', 'Mins']]

df['Mins'] = df['Mins'].str.replace(r"[-]", '0')
df['Mins'] = df['Mins'].str.replace(r"[.'-]", '')
df['Age'] = df['Age'].astype(int)
df['Mins'] = df['Mins'].astype(int)
df['MinsPer%'] = 100 * df['Mins'] / Mins_Played
df['Load'] = df['Age']*df['Mins']
#df = df[df['Mins'] > 0].reset_index(drop = True) # Drop players with 0 mins played
df
```

```python
df_domestic = df[df['Nationality'] == 'Turkey']
Domestic_Bias = round(100*sum(df_domestic['Mins']) / sum(df['Mins']), 1)
print('Domestic Player Bias %: ' + str(Domestic_Bias))
```

```python
Average_Age = round(df['Age'].mean(), 1)
Weighted_Average_Age = round(sum(df['Load']) / sum(df['Mins']), 1)
print('Squad Average Age: ' + str(Average_Age))
print('Squad Weighted Average Age: ' + str(Weighted_Average_Age))
```

```python
def plotParams():
    plt.title(f'{Team_Name} Age vs. Minutes Played', color='white', fontweight='bold')
    plt.xlabel('Age', color='white', fontweight = 'bold')
    plt.ylabel('Minutes Played', color='white', fontweight='bold')
    ax.set_facecolor('#121212')
    fig.set_facecolor('#121212')
    ax.tick_params(color='white', labelcolor='white', length=0)
    for spine in ax.spines.values():
        spine.set_edgecolor('white')
        spine.set_visible(False)
    plt.xticks(range(16, 41, 4))
    plt.gca().set_xlim(left=15.5)
    plt.gca().set_xlim(right=40.5)
    plt.grid(b=True, which='major', color='white', zorder=0, lw=1, ls='--', alpha=0.25)
    plt.rcParams['figure.dpi'] = 300
```

```python
fig, ax = plt.subplots(figsize = (12,9.6))
#ax.scatter(df['Age'], df['Mins'], marker='o')
g = sns.scatterplot(data=df, x='Age', y='Mins',
                    hue='Nationality', palette='Paired', marker='o', s=50, legend=False, zorder=2)
plotParams()
plt.yticks(range(0, Mins_Played, 500))
extraticks = [Mins_Played]
plt.yticks(list(plt.yticks()[0]) + extraticks)
plt.gca().set_ylim(bottom=-180)
plt.gca().set_ylim(top = 1.05 * Mins_Played)
plt.axvline(x=Weighted_Average_Age, color='#ff5454', zorder=0, lw=2, alpha=0.75)
plt.text(0.972*Weighted_Average_Age, -90,
         Weighted_Average_Age, rotation=0, color='#ff5454', zorder=0, alpha=1, size=10)
plt.text(0.958*Weighted_Average_Age, -215,
         'Weighted \n Avg. Age', rotation=0, color='#ff5454', zorder=0, alpha=1, size=7)
plt.text(1.005 * 16, 1.005 * Mins_Played,
         'Domestic Player Bias: ' + str(Domestic_Bias) + '%', rotation=0, color='#ff5454', zorder=0, alpha=1, size=8)
#plt.axhline(y=df['Mins'].mean(), color='red', zorder=1, lw=1, alpha=0.25)

# label points on the plot
for x, y, z in zip(df['Age'], df['Mins'], df['player']):
    plt.text(x=x+0.15,  # x-coordinate position of data label
             y=y-20,  # y-coordinate position of data label, adjusted to be 150 below the data point
             s=z,  # data label, formatted to ignore decimals
             color='white', size=7, zorder=2)  # set colour of line
fig.tight_layout()
fig.show()
```
